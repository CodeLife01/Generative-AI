### âœ… Step 2: Start Installation

1. Locate the downloaded `.exe` file (usually in the **Downloads** folder).

### âœ… Step 3: Setup Wizard

1. **Welcome Screen**
   Click **Next** to proceed.

2. **License Agreement**
   Read and accept the license, then click **Next**.

3. **Choose Components**

   * Leave default components selected (recommended).
   * Ensure "**TShark**" and "**WinPcap/Npcap**" are selected.
     Click **Next**.

4. **Select Shortcuts**
   Choose whether you want desktop shortcuts. Click **Next**.

---

### âœ… Step 4: Install Npcap (Packet Capture Driver)

1. Npcap is required for capturing packets.
2. Click **Install** when the Npcap installer appears.
3. Check options like:

   * **Install Npcap in WinPcap API-compatible Mode** (recommended)
4. Click **Next**, then **Install**.

---

### âœ… Step 5: Complete Wireshark Installation

1. Once Npcap is installed, the main Wireshark installer continues.
2. Click **Next** then **Install**.
3. Wait for the installation to finish.
4. Click **Finish**.

---

### âœ… Step 6: Launch Wireshark

1. Open **Wireshark** from the **Start Menu** or **Desktop Shortcut**.
2. On the main screen, you'll see a list of network interfaces.
3. Double-click any interface to start capturing packets.
