### OLLAMA

Ollama supports embedding models, making it possible to build Retrieval Argumented Generation (RAG) applications that combine text prompts with existing documents or other data.


```python
from langchain_ollama import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model="gemma3:1b") ## By default it uses llama2 
)
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpha is the first letter of Greek alphabet",
        "beta is the second letter of Greek alphabet"
    ]
)
```

```python
embeddings.embed_query("what is the second letter of Greek alphabet")
```

#### gemma3 don't support embedding


```python
## Other Embedding Models
embeddings = (OllamaEmbeddings(model="mxbai-embed-large") ## By default it uses llama2
              )
text = "This is a text document."
query_result = embeddings.embed_query(text)
```
